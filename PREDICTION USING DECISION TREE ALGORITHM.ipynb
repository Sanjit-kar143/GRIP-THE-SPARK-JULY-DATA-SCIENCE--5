### FEATURE ENGINEERING 

```python
# we take difference of each feature with each other get more information between them
data['SepalDiffCm'] = data['SepalLengthCm']-data['SepalWidthCm']
data['PetalDiffCm'] = data['PetalLengthCm']-data['PetalWidthCm']
data
```

```python
data['SepalPetalLenDiff'] = data['SepalLengthCm'] - data['PetalLengthCm']
data['SepalPetalWidthDiff'] = data['SepalWidthCm'] - data['PetalWidthCm']
data['SepalPetalLenWidthDiff'] = data['SepalLengthCm'] - data['PetalWidthCm']
data['SepalPetalWidthLenDiff'] = data['SepalWidthCm'] - data['PetalLengthCm']
data
```

```python
sns.pairplot(data[['Species','SepalDiffCm','PetalDiffCm','SepalPetalLenDiff','SepalPetalWidthDiff','SepalPetalLenWidthDiff','SepalPetalWidthLenDiff']],hue= 'Species')
```

### BULIDING A CLASSIFICATION MODEL 

```python
#import important libaries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
mod = DecisionTreeClassifier(criterion='entropy') 
```

```python
x = data.drop('Species',axis=1)
y = data['Species']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state = 0)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
model1 = mod.fit(X_train,Y_train)
```

```python
y_pred = mod.predict(X_test)
```

```python
print(classification_report(y_pred,Y_test))
```

```python
# see the Accuracy
accuracy_score(y_pred,Y_test)
```
